## TASK 13

```python
""" Creates histogram for each grade """
mssGrade = schools.groupby('Grade')['Mean Scale Score']
for grade, i in mssGrade:
    plt.hist(i, alpha=0.5)
    plt.xlabel('Mean Scale Score')
    plt.ylabel('Frequency')
    plt.title(f'Mean Scale Score for Grade {grade}')
    plt.show()
```

## TASK 14

```python
""" Creates scatter plot for district """
mssDistrict = schools.groupby('District')['Mean Scale Score']
mssDistrict.plot(style='o')
plt.xlabel('District')
plt.ylabel('Mean Scale Score')
plt.title('Mean Scale Score by District')
plt.show()
```

## TASK 27

```python
""" Creates subset (grade 3 only, year 2012 only) of schools """
schools_subset = schools.loc[(schools['Grade'] == '3') & (schools['Year'] == 2012), ['District', 'Grade', 'Year', 'Mean Scale Score']]
schools_subset
```

## TASK 28

```python
""" Function creates rank pertaining to MSS for each district """
def default_rank(df, column='Mean Scale Score'):
    df['Default Rank'] = df[column].rank(ascending=False)
    return df
```

```python
""" Displays schools_subset (grade 3, year 2012) df before and after passing the function """
print("Before adding rank:\n", schools_subset, "\n\n")
schools_subset = default_rank(schools_subset)
print("After adding rank: \n", schools_subset, "\n")
```

## TASK 29

```python
""" Displays schools2 using df schools_subset (grade 3 and year 2012)"""
print("Displays schools2 using df schools_subset:")
schools2 = schools_subset.groupby(['Grade', 'Year']).apply(default_rank)
schools2.sort_values(by= 'Default Rank', ascending=True)
```

```python
"""Displays schools3 using df schools (all grade and all years)"""
print("Displays schools3 using df schools:")
schools3 = schools.groupby(['Grade', 'Year']).apply(default_rank)
schools3.sort_values(by= 'Default Rank', ascending=True)
```
